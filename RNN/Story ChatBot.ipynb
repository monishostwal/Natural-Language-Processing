```python
vocab=[]
for story,question,answer in all_data:
    story_unique_words=list(set(story))
    for word in story_unique_words:
        if(word not in vocab):
            vocab.append(word)
    question_unique_words=list(set(question))
    for word in question_unique_words:
        if(word not in vocab):
            vocab.append(word)
```

```python
import keras
```

```python
from keras.preprocessing.sequence import pad_sequences
from keras.preprocessing.text import Tokenizer
```

```python
tokenizer=Tokenizer(filters=[])
```

```python
tokenizer.fit_on_texts(vocab)
```

```python
tokenizer.word_index # It also done lower case automatically
```

```python
x=[]
x_q=[]
a=[]
for story,question,answer in train_data:
    xc=[]
    for word in story:
        
        xc.append(tokenizer.word_index[word.lower()])
    x.append(xc)
    xac=[]
    for word in question:
        xac.append(tokenizer.word_index[word.lower()])
    x_q.append(xac)
    y=np.zeros(len(tokenizer.word_index)+1)
    y[tokenizer.word_index[answer]]=1
    a.append(y)
```
